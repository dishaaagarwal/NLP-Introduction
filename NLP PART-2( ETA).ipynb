```python
#link-https://github.com/adashofdata/nlp-in-python-tutorial/blob/master/2-Exploratory-Data-Analysis.ipynb

import pandas as pd
```

```python
data=pd.read_pickle('dtm.pkl')
data=data.transpose()
data.head()
```

```python
#data['ali'].sort_values(ascending=False).head(30)
top_word_dict={}
for i in data.columns:
    top=data[i].sort_values(ascending=False).head(30)
    top_word_dict[i]=list(zip(top.index,top.values))
```

```python
#top_word_dict
```

```python
for comedian,top_word in top_word_dict.items():
    print(comedian)
    print(',  '.join([word for word,count in top_word[0:14]]))
    print('--------------')
```

```python
from collections import Counter

word_list=[]
for comedian in data.columns:
    #print(comedian)
    #print( [word for word,count in top_word_dict[comedian]])
    top=[word for word,count in top_word_dict[comedian]]
    for t in top:
        word_list.append(t)
```

```python
word_list
```

```python
Counter(word_list).most_common()
```

```python
add_stop_word=[]
add_stop_word=[ w for w,c in Counter(word_list).most_common() if c>6]
add_stop_word
```

```python
from sklearn.feature_extraction import text 
from sklearn.feature_extraction.text import CountVectorizer

# Read in cleaned data
data_clean = pd.read_pickle('data_clean.pkl')
#data_clean
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_word)

# Recreate document-term matrix
cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(data_clean.transcript)
data_stop = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_stop.index = data_clean.index

# Pickle it for later use
import pickle
pickle.dump(cv, open("cv_stop.pkl", "wb"))
data_stop.to_pickle("dtm_stop.pkl")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#wordcloud = WordCloud().generate(text)
wc = WordCloud(stopwords=stop_words, background_color="white", colormap="Dark2",
               max_font_size=150, random_state=42)



plt.rcParams['figure.figsize'] = [16, 6]

full_names = ['Ali Wong', 'Anthony Jeselnik', 'Bill Burr', 'Bo Burnham', 'Dave Chappelle', 'Hasan Minhaj',
              'Jim Jefferies', 'Joe Rogan', 'John Mulaney', 'Louis C.K.', 'Mike Birbiglia', 'Ricky Gervais']
```

```python
for index,comedian in enumerate(data.columns):
    #print(index,comedian)
    wc.generate(data_clean.transcript[comedian])
    
    plt.subplot(3,4,index+1)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(full_names[index])
plt.show()    
```

# Number Of Words Analysis


```python
unique_list=[]
for comedian in data.columns:
    #print(comedian,data[comedian].nonzero()[0].size)
    unique=data[comedian].nonzero()[0].size
    unique_list.append(unique)
#unique_list    


data_words = pd.DataFrame(list(zip(full_names,unique_list)),columns=['Full_Name','Unique_Words']   )
data_unique_sort=data_words.sort_values(by='Unique_Words')
data_unique_sort
```

```python
total_word_list=[]
for comedian in data.columns:
    #print(comedian,sum(data[comedian]))
    total=sum(data[comedian])
    total_word_list.append(total)
    
    
run_times = [60, 59, 80, 60, 67, 73, 77, 63, 62, 58, 76, 79]  


data_words['Total_Words']=total_word_list
data_words['Run_Time']=run_times
data_words['Words_per_Minute']= data_words['Total_Words']/data_words['Run_Time']
data_wpm_sort=data_words.sort_values(by='Words_per_Minute')
data_wpm_sort
```

# Amount of Profanity

```python
#Counter(word_list).most_common()
```

```python
data_bad_words = data.transpose()[['fucking', 'fuck', 'shit']]
data_profanity = pd.concat([data_bad_words.fucking + data_bad_words.fuck, data_bad_words.shit], axis=1)
data_profanity.columns = ['f_word', 's_word']


data_profanity
```

```python
plt.rcParams['figure.figsize'] = [10, 8]

for i, comedian in enumerate(data_profanity.index):
    x = data_profanity.f_word.loc[comedian]
    y = data_profanity.s_word.loc[comedian]
    plt.scatter(x, y, color='blue')
    plt.text(x+1.5, y+0.5, full_names[i], fontsize=10)
    plt.xlim(-5, 155) 
    
plt.title('Number of Bad Words Used in Routine', fontsize=20)
plt.xlabel('Number of F Bombs', fontsize=15)
plt.ylabel('Number of S Words', fontsize=15)

plt.show()
```
